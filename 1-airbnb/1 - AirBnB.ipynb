```python
import pandas as pd
import numpy as np
from bokeh.charts import Bar, output_file, show
from bokeh.io import output_notebook
from bokeh.resources import CDN
from bokeh.embed import file_html
```

```python
sfdf = pd.read_csv('http://data.insideairbnb.com/united-states/ca/san-francisco/2015-09-02/visualisations/listings.csv')

sfdf['city'] = 'SF'
sfdf
```

```python
nycdf = pd.read_csv('http://data.insideairbnb.com/united-states/ny/new-york-city/2015-09-01/visualisations/listings.csv')
mandf = nydf[nydf['neighbourhood_group'] == 'Manhattan']

mandf['city'] = 'NYC'

mandf
```

```python
joineddf = pd.concat([mandf, sfdf])

joineddf
```

```python
TOOLS = 'crosshair,resize,save'

p = Bar(joineddf, label='room_type', values='price', agg='median', group='city',
        title="San Francisco is more expensive and antisocial than Manhattan", legend='top_right', tools=TOOLS, ylabel = 'Median Price', xlabel = 'Room Type')


#output_notebook()
#output_file("1-airbnb.html")
#show(p)

html = file_html(p, CDN, "1-airbnb")

text_file = open("1-airbnb.txt", "w")
text_file.write(html)
text_file.close()

```
